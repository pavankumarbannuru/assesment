```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import expon

# Simulate customer arrival times using exponential distribution
rate_parameter = 1/5  # Average arrival every 5 minutes
arrival_times = np.random.exponential(scale=1/rate_parameter, size=1000)

# Plot the simulated arrival times
plt.hist(arrival_times, bins=30, density=True, alpha=0.7, color='blue')

# Plot the probability density function (PDF) of exponential distribution
x = np.linspace(0, 30, 100)
pdf = expon.pdf(x, scale=1/rate_parameter)
plt.plot(x, pdf, 'r-', lw=2, label='Exponential PDF')

plt.title('Customer Arrival Times at a Bank')
plt.xlabel('Time Between Arrivals (minutes)')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```
