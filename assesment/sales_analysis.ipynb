```python
count=df.groupby('SKU_Category')['Sales_Amount'].count().sort_values(ascending=False)
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=count.index,y=count.values)
plt.rcParams.update({'font.size':10})
plt.ylabel('count')
plt.xticks(rotation='vertical')
plt.show()
```

```python
for i,j in zip(count.index,count.values):
    if j>100:
        print(i,j)
```

```python
sales_groupby=df.groupby('SKU_Category')['Sales_Amount'].sum().sort_values(ascending=False)
sales_groupby
```

```python
plt.figure(figsize=(10,10))
sns.barplot(sales_groupby.index, sales_groupby.values)
plt.xlabel('SKU Category')
plt.ylabel('Sales Amount Summation')
plt.title('Sales Amount Summation for Each SKU Category')
custom_tick_positions = [idx for idx, val in enumerate(sales_groupby.index) if idx % 4 == 0]
custom_tick_labels = [sales_groupby.index[i] for i in custom_tick_positions]
plt.xticks(custom_tick_positions, custom_tick_labels, rotation=45)  # Use custom tick positions and labels
plt.show()
```

```python
for i,j in zip(sales_groupby.index,sales_groupby.values):
    if j>100:
        print(i,int(j))
```
